```
clinicaltrial_2021="/FileStore/tables/clinicaltrial_2021.csv" #file location for clinicaltrial_2021 after uploading datasets
mesh="/FileStore/tables/mesh.csv"; #file location for mesh file after uploading datasets
pharma="/FileStore/tables/pharma.csv"
##file location for pharma file after uploading datasets
```

```
clinicrdd= sc.textFile(clinicaltrial_2021)  # read from text file words.txt on DBFS
clinicrdd.take(5)
```

```
header = clinicrdd.first() #declared that header is the first row of the RDD 
clinicrdd1=clinicrdd.filter(lambda line: line != header) #filter() function was used to filter the rows from RDD that is not header (first row) to return a new RDD
clinicrdd1.take(5) #take and display five rows of the new RDD
```

```
meshrdd= sc.textFile(mesh)   # read from text file words.txt on DBFS 
meshrdd.collect()
```

```
meshheader = meshrdd.first()
meshrdd1=meshrdd.filter(lambda line: line != meshheader)
meshrdd1.collect()
```

```
splrdd= clinicrdd1.map(lambda x: x.split("|")) #map() was used to apply lambda(transformation function) on every element of clinicrdd1 using by split () function used for splitting the string column on ("|") delimeter
splrdd.take(10)
```

```
import re
splitRDD=meshrdd1.map(lambda myrdd: re.split(',',myrdd)) #re.split() function used so that a list of string with matching occurrence of pattern in the string separated by ","
splitRDD.collect()
```

```
pharmardd= sc.textFile(pharma)  # read from text file words.txt on DBFS
pharmardd.collect()
```

```
pharmaheader = pharmardd.first()
pharmardd1=pharmardd.filter(lambda line: line != pharmaheader)
pharmardd1.collect()
```

```
splpharmardd= pharmardd1.map(lambda x: x.split(",")) #map() was used to apply lambda(transformation function) on every element of clinicrdd1 using by split () function used for splitting the string column on (",") delimeter
splpharmardd.take(10)
```

```
print ("The number of studies for clinicaltrial_2021 is", clinicrdd1.count())
```

```
type2= splrdd.map(lambda x: (x[5],1)).reduceByKey(lambda x1,x2: x1+x2) 
Type= type2.sortBy(lambda a: -a[1]) 
Type.collect()
```

```
conditionrdd= clinicrdd1.map(lambda x: x.split("|")).map(lambda s: (s[7]).split(','))
skipspace=[" ", ""]
remnull=conditionrdd.filter(lambda l:l not in skipspace)
condition=remnull.flatMap(lambda x:x).filter(lambda x:len(x)>0)  # flatMap() to output multiple elements for each input value, split on space
condition2= condition.map(lambda x: (x,1)) # Map a tuple and append int 1 for each word in words.txt
condition3=condition2.reduceByKey(lambda x1,x2: x1+x2) # Perform aggregation (sum) all the int values for each unique key
Topconditions=condition3.sortBy(lambda ascend:-ascend[1])
Topconditions.take(5)
```

```
splmesh=meshrdd1.map(lambda x:(x.split(",")[-1],x.split(',')[:-1]))
splmeshdata=splmesh.flatMapValues(lambda s:s)\
    .map(lambda s:(s[1],s[0].split(".")[0]))  # flatMap() to output multiple elements for each input value, split on (".")
splmeshdata.take(10)
```

```
splclinicalrdd=clinicrdd1.map(lambda s:s.split('|')).map(lambda s:(s[0],s[7].split(',')))
flatsplclinicalrdd=splclinicalrdd.map(lambda x:(x[0],x[1]))\
    .flatMapValues(lambda x:x)\
    .map(lambda x:(x[1],x[0]))\
.filter(lambda x:(len(x[0])>0))
flatsplclinicalrdd.take(10)
```

```
frequentroots =flatsplclinicalrdd.join(splmeshdata).values()
toproots = frequentroots.map(lambda s: (s[1],1))\
.reduceByKey( lambda v1,v2:v1+v2)\
.sortBy(lambda a: -a[1])
toproots.take(5)
```

```
splpharmardd= pharmardd1.map(lambda x:(x.split('","')[1],1))
clinicsponsor = clinicrdd1.map(lambda x: x.split("|"))\
    .map(lambda x:(x[1],1))
Sponsors=clinicsponsor.subtract(splpharmardd)\
    .map(lambda x:(x[0],1))\
.reduceByKey(lambda x1,x2: x1+x2)\
.sortBy(lambda a:-a[1])
Sponsors.take(10)
```

```
completed = splrdd.map(lambda x:(x[2],x[4])).filter(lambda x:x[0]=="Completed")
completion=completed.filter(lambda x:len(x[1])>0)\
    .filter(lambda x:x[1].split(" ")[1]=="2021")\
    .map(lambda x:(x[1].split (" ")[0],1))\
    .reduceByKey(lambda x1,x2:x1+x2)
completion.sortByKey().collect()
```

```
import calendar
c = {i:e for e,i in enumerate(calendar.month_abbr[1:],1)}

#{'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6, 'Jul': 7, 
#'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12}

Completion=completion.sortBy(keyfunc=lambda x: c.get(x[0]))
Completion.collect()
```
